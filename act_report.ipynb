# Data analysis and visualization report on WeRateDogs tweet archive

## Introduction:

In this project, I intent to analyze the tweet archive of Twitter user [@dog_rates](https://twitter.com/dog_rates), also known as [WeRateDogs](https://en.wikipedia.org/wiki/WeRateDogs). WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. WeRateDogs asks people to send photos of their dogs, then tweets selected photos with a rating. Dogs are rated on a scale of one to ten, but are invariably given ratings in excess of the maximum, such as "13/10". 

Every image in the tweet archive file is run through a neural network that can identify breeds of dogs and the top 3 predictions identified by the algorithm are included in the file for each tweet. The file is then cleaned and is used as an input for this analysis. 

Note: Only the original tweets that has a rating for a dog are included in the input file and used for analysis.

## Analysis #1: How does the rating statistics look like?

Here is the descriptive statistics for the dog ratings:

There are 1685 ratings with a mean rating of 10.8 and standard deviation of 1.8. The ratings range from 0 to 14. Even though the ratings are provided out of ten, the mean and the median ratings are above ten. This indicates that more dogs are rated above ten. Let's confirm this by looking at the actual numbers per rating. 

Out of 1685 dogs that were rated, 1043 dogs or 62% received a rating higher than 10. Let's look at the histogram of the rating distribution and see how the higher ratings skewed the distribution.

As seen from the above graph, the distribution of dog rating is left skewed.

## Insight #1:
The ratings range from 0 to 14 with a mean of 10.8 and standard deviation of 1.8. With a median rating of 11, which is slightly higher than the mean, the distribution of ratings is left skewed. The left skewness is also confirmed by the more number of dogs rated on the higher end of the scale (62% of the dogs received a rating of 11 or higher).
